```python
import numpy as np
from lightkurve import *
import matplotlib.pyplot as plt
import lightkurve as lk
print('good')

```

```python
TIC = 'TIC 284475976' #Confirmed Planet WASP-48b
#search what data is available for a given target
sector_data = lk.search_lightcurve(TIC)
sector_data
#sector_data=lk.search_lightcurve(TIC,author='SPOC',	exptime=120)
```

```python
lc = sector_data[1].download() 
lc.plot(linewidth= 0, marker = '.', color = 'midnightblue', alpha = 0.5)
```

```python
lc_norm=lc.normalize()
lc_norm.plot(linewidth= 0, marker = '.', color = 'midnightblue', alpha = 0.5)
```

```python
#lc_clean = lc_norm.remove_nans().remove_outliers()
t0 = 2825.459527
period = 2.14363
lc_fold = lc_norm.fold(period = period, epoch_time = t0)
lc_fold.plot(linewidth =0, color = 'darkviolet', marker='.', alpha=0.5 )
#plt.xlim(-2,2)
#print(lc_norm.time)
```

```python
lc_phased_binned = lc_fold.bin (5/24/60) #bin width =15 mins. it has to be in terms of minutes when using Lightkurve
fig, ax = plt.subplots (figsize = (7,5))
lc_fold.plot(ax = ax, marker = '.', linewidth = 0, color = 'blue', alpha = 0.2, markersize = 3, label = "unbinned")
lc_phased_binned.plot(ax= ax, marker = 'o', linewidth= 0, color = 'gold', alpha = 1, markersize = 3, label = "binned")
plt.xlim(-0.5,0.5)
```
